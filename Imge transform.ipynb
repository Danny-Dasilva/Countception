```python
from PIL import Image
import numpy
im = Image.open(label)
np_im = numpy.array(im)
print(np_im.shape)
```

```python
import numpy as np

# for color image
color = True
pixels = np.argwhere(np_im == color)
```

```python
pixels
```

```python
l_pix = pixels.tolist()
```

```python
l_pix
```

```python
import cv2
import numpy as np

width = 600
height = 600

# Make empty black image of size (100,100)
img = np.zeros((height, width, 3), np.uint8)

red = [0,0,255]

# Change pixel (50,50) to red

for x, y in l_pix:
    #cv2.circle(img,(x, y), 5, (0,255,0), -1)
    img[x, y] = red
    
    
    
cv2.imshow('img', img)
cv2.waitKey(0)
```
